# Visualising titanic survival using seaborn
The discription for the dataset can be found here - https://www.kaggle.com/c/titanic/data


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as numpy
import seaborn as sns
```

```python
titanic = pd.read_csv('/Users/Suha/Projects/TestGitHub/Titanic/train.csv')
```

```python
titanic.head()
titanic
```

```python
titanic.describe()
```

```python
cols = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
titanic = titanic[cols].dropna()
titanic
# from 891 rows the dataset now has 183 rows only 
#As this is just a visualisation project we will just work with the current numbers
```

```python
# lets plot the smoothed out histogram using seaborn distplot() method. This does use the matplotlib under the hood for the functions
# note that it automatically scales the axis and everything
sns.distplot(titanic['Fare'])
plt.show()

sns.distplot(titanic['Age'])
plt.show()
```

```python
# to just make the kde plot we will use kedplot() method
sns.kdeplot(titanic['Age'])
plt.show()
```

```python
# Adding shading to the plot and other style elements
sns.set_style("whitegrid")
sns.kdeplot(titanic['Age'], shade = True)
sns.despine(left = True, bottom = True)
plt.xlabel('Age')
plt.show()
```

    We will use Facetgrid object in the seaborn library to plot 2 different KDE for the age distribution of survived pasengers and the passenders that have not survived. The input to the factgrid is either a matplotlib or a seaborn object

```python
f = sns.FacetGrid(titanic , col = "Survived", size = 6)
f.map(sns.kdeplot, 'Age')
plt.show()
```

```python
f = sns.FacetGrid(titanic , col = "Survived", size = 6)
f.map(plt.hist, 'Age')
plt.show()
```

```python
f = sns.FacetGrid(titanic , col = "Survived", size = 6)
f.map(sns.distplot, 'Age')
plt.show()
```

    Let us now create KDE for each of the 3 categories of pasenger class by age

```python
p = sns.FacetGrid(titanic, col = 'Pclass', size = 6)
p.map(sns.kdeplot, "Age")
sns.despine(left = "off", bottom = 'off')
plt.show()
```

```python
p = sns.FacetGrid(titanic, col = 'Pclass', size = 6)
p.map(sns.distplot, "Age")
sns.despine(left = "off", bottom = 'off')
plt.show()
```

    Let us try to base the plot on two conditions now

```python
t = sns.FacetGrid(titanic, col = "Survived", row = 'Pclass', size = 6 )
t.map(sns.distplot, 'Age').add_legend('male', 'female')
sns.despine(left = 'off', bottom = 'off')
plt.show()
```

```python
t = sns.FacetGrid(titanic, col = "Survived", row = 'Pclass',hue = 'Sex', size = 3 )
t.map(sns.kdeplot, 'Age', shade = True).add_legend()
sns.despine(left = 'off', bottom = 'off')
plt.show()
```
