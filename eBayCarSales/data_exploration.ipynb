Lets have a look at the brand columns. We will find out some descriptive statistics for top 20 brands by count

```python
top_brands = autos_filter["brand"].value_counts(ascending = False).head(20).keys()
print(top_brands)

brand_mean_price = {}

#create an empty dictionary to store the mean values 

for brand in top_brands:
    brand_mean_price[brand] = autos_filter.loc[autos_filter["brand"] == brand,"price"].mean()

sorted(brand_mean_price.items(), key = lambda kv:kv[1],reverse = True)
```

As seen from the above analysis, the  top 3 most listed used car in the classified are german in make. Not a huge surpirse as this is a german classified.The brand feching the most value also belongs to volkswagon group. The one car which is listed more and also fetches more price is mercedes_benz. Ford falls has average listin and price in the top 20 brands

Now we will check if the mean price has any relationship with the mean milage


```python
brand_mean_milage= {}
for brand in top_brands:
    brand_mean_milage[brand] = autos_filter.loc[autos_filter["brand"] == brand,"kilometer"].mean()


print(brand_mean_milage)
```

in order to put together the milage and the price, we will convert each of these dictionaries as  series 
and then to a dataframe object

```python
milage = pd.DataFrame(pd.Series(brand_mean_milage), columns = ["milage"])

milage["price"] = pd.Series(brand_mean_price)

milage.sort_values("milage", ascending = False)
```

sonstige_autos might have fetched more price as the average milage was less. On the other hand, BMW seems to fetch more despite of high milage. So does audi and volkswagen
